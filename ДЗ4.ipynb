## Условие 2: 2 задача

### 2.1. Изучите распределение домов от наличия вида на набережную
Постройте график

Сделайте выводы



```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

### 2.2. Изучите распределение этажей домов



```python
data = df['floors'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

### 2.3. Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Цена дома изменяется в зависимости от жилой площади. Чем выше площадь, тем выше цена.

```python
plt.figure(figsize=(6,4))

plt.scatter(df['yr_built'], df['price'])

plt.title('График разброса значений')
plt.xlabel('yr_built')
plt.ylabel('price')
```

Не наблюдается прямой корреляции между годом постройки и ценой недвижимости

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('цена')
plt.ylabel('количество этажей')
plt.title('Распределение цены от количества этажей в доме');
```

```python
Максимальную цену достигают дома с количеством этажей от 2 до 3.
```

```python
plt.figure(figsize=(6,4))

plt.bar(df['bedrooms'], df['price'])

plt.title('Зависимость цены от количества спален')
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.xticks(rotation=30)
```

```python
Максимальная цена на дома с количеством спален 6.
```
